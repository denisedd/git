# Scrape the Guardian News

```python
import time
import os, sys
import requests
import json
from datetime import datetime
import pandas as pd

def json_html(parameters):
    response = requests.get("https://content.guardianapis.com/search", params=parameters)
    data = json.loads(response.text)
    try:
        content = data['response']['results']
    except:
        pass
    return data, content
    
def html_txt(data, parameters, lst, path, size=50):
    for i in range(size):
        try:
            url = data['response']['results'][i]['webUrl']
            response = requests.get(url)
            file = url.split('https://')[1]
            file1 = file.replace('/',':')
            lst.append(file1)
            filename = path+file1
            f = open(filename,'w') 
            f.write(response.text)
            f.close()
        except IndexError:
            pass
        continue
            
    return lst

    
def Guardian_API(query, path, start_date, end_date):
    lst=[]
    array=[]
    parameters = {}
    parameters['api-key'] = 'b0407d2c-612a-48d8-81a8-2a4166752b4d'
    parameters['q']=query
    parameters['page']=1
    parameters['page-size']=50
    parameters['section'] ='business'
    parameters['from-date']=''
    parameters['to-date']=''
    parameters['show-fields']='all'
    parameters['order-by']='oldest'
    
    parameters['from-date'] = start_date
    parameters['to-date'] = end_date
        
    data, content = json_html(parameters)
    ttl_message = data['response']['total']
    print(ttl_message)
    if ttl_message==0:
        return 
    total_pages = data['response']['pages']
    
    for i in range(1, total_pages):
        parameters['page'] = i
        data, content = json_html(parameters)
        lst=html_txt(data, parameters, lst, path)
        array.append(content)
        
        if i%9==0:
            time.sleep(60)
    parameters['page'] = total_pages
    data, content = json_html(parameters)
    array.append(content)
    lst=html_txt(data, parameters, lst, path, ttl_message%50)
    filename = path+'index.json'
    with open(filename,'w') as outfile:
        json.dump(array, outfile, sort_keys=True, indent=4, separators=(',',':'))
    
query =['U.S finance', 'U.S stocks', 'U.S stocks market', 'U.S economy', 'Apple Inc.', 'Alibaba', 'Amazon', 'Google', 'Facebook', 'Netflix', 
                'Microsoft', 'Tesla', 'eBay', 'Cisco']


date_Dict = {}
df = pd.date_range(start='1/1/2016', periods=12, freq='M')
for date in df:
    end_date = datetime.strftime(date, '%Y-%m-%d')
    date = end_date.split('-')[:-1]
    date.append('01')
    start_date = '-'.join(a for a in date)
    date_Dict[start_date] = end_date
    
for key, value in date_Dict.items():
    for a in query:
        date=datetime.strptime(key, '%Y-%m-%d')
        month = date.strftime('%m')
        path ='../../Data/The Guardian/2016/{}/{}/'.format(month, a)
        os.makedirs(path)
        print(key)
        print(value)
        Guardian_API(a, path, key, value)
        time.sleep(15)
```
