# Data Extraction on Guardian news
1. From the returned api Guardian news dataset, we only get the dataset with published date, headline, and url of news.
2. With beautiful soup, scrape the informative content of the whole news article

```python
from bs4 import BeautifulSoup
import os
import pandas as pd

folders = ['U.S economy/', 'U.S finance/', 'U.S stocks/', 'U.S stocks market/']

headline_lst = []
date_lst = []
header_lst = []
all_content_lst = []

for month in range(1, 13):
    print('reading {} month data'.format(month))
    if month<10:
        month = str(0)+str(month)
    for folder in folders:
        path = os.path.join('../../Data', 'The Guardian', '2018', str(month), folder)
        for file in os.listdir(path):
            if not file.startswith('.'): # filter hidden file
                with open(path+file,'r') as news:
                    data = news.read()
#                     print(file)
                soup = BeautifulSoup(data, features="html.parser")
                try:
                    content = soup.find("div", {"class": "content__article-body from-content-api js-article__body"})
                    content_rr = content.find_all('p')
                    content_lst = [a.text for a in content_rr]
                    all_content = ' '.join(content_lst)
                    headline = soup.find("h1", {"class": "content__headline"}).get_text()
                    time = soup.find("time", {"itemprop": "datePublished"}).get_text()
                    time_date = ' '.join(time.split(' ')[1:-1])
                    header = soup.find("div", {"class": "content__standfirst"})
                    try:
                        header_rr = header.find('p').get_text()
                    except:
                        try:
                            header_rr = header.find_all('li')
                            heads_lst = [a.text for a in header_rr]
                            all_headers = ' '.join(heads_lst)
                        except:
                            header_rr=''

                    headline_lst.append(headline)
                    date_lst.append(time_date)
                    header_lst.append(header_rr)
                    all_content_lst.append(all_content)

                except:
                    headline = soup.find("h1", {"class": "content__headline"}).get_text()
                    time = soup.find("time", {"itemprop": "datePublished"}).get_text()
                    time_date = ' '.join(time.split(' ')[1:-1])
                    try:
                        header = soup.find("div", {"class": "content__standfirst"})
                        header_rr = header.find('p').get_text()
                        header_lst.append(header_rr)
                    except:
                        header_lst.append("")
                    content = soup.find("div", {"class": "content__standfirst"})
                    content_rr = content.find_all('li')
                    content_lst = [a.text for a in content_rr]
                    all_content = ' '.join(content_lst)

                    headline_lst.append(headline)
                    date_lst.append(time_date)
                    all_content_lst.append(all_content)

                    main = soup.find("div", {"class": "content__main tonal__main tonal__main--tone-dead"})
                    headers = main.find_all('div', {"class":["block block--content", "block block--content is-key-event"]})

                    for a in headers:
                        try:
                            headline1 = a.find("h2", {"class": "block-title"}).get_text()
                            headline_lst.append(headline1)
                        except:
                            headline_lst.append("")
                        cnt = a.find("div", {"class": ["block-elements block-elements--no-byline", "block-elements"]})
                        cnt1 = cnt.find_all('p')
                        cnt1_lst = [a.text for a in cnt1]
                        all_cnt = ' '.join(cnt1_lst)
                        all_content_lst.append(all_cnt)
                        date_lst.append(time_date)
                        header_lst.append('')
    
Guard_news = pd.DataFrame({
            'headline': headline_lst,
            'date': date_lst,
            'header': header_lst,
            'news': all_content_lst
 })
Guard_news['headline'] = Guard_news['headline'].str.strip('\n')
Guard_news['header'] = Guard_news['header'].str.strip('[<li>')
Guard_news['content'] = Guard_news['header']+Guard_news['news']
Guard_news['date'] = pd.to_datetime(Guard_news['date'])
Guard_news
# Export = Guard_news.to_json ('Data/processed_data/Guardian/Guardian_2018.json')

```
