# Text Preprocessing
1. Load Json News from Guardian and New York Times.
2. Remove extra whitespace.
3. Tokenize.
4. Remove punctuation, stopwords, convert to lower case.
5. Lemmatize.

```python
import pandas as pd
import json
import os
```

```python
# Read Guardian News.
Guardian_df = pd.DataFrame(columns=['headline', 'date', 'header','news','content'])
root = '../../Data/JsonNews/Guardian/General'
for f in os.listdir(root):
    if f.endswith('.json'):
        with open(os.path.join(root, f)) as file:
            Guardian_df = pd.concat([Guardian_df, pd.read_json(file)])  
            
# Keep headline, date, content
Guardian_df = Guardian_df.drop(['header', 'news'], axis=1)
Guardian_df
```

```python
# Read New York Times News.
NYT_df = pd.DataFrame(columns=['headline', 'date', 'header','news','content'])
headlines = []
dates = []
contents = []
root = '../../Data/JsonNews/NYT'
for year in os.listdir(root):
    if year.isdigit():
        for month in os.listdir(os.path.join(root, year)):
            if month.isdigit():
        #         print('Year:' , year, 'Month:', month)
                for file in os.listdir(os.path.join(root, year, month)):
                    if file.endswith('.json'):
                        with open(os.path.join(root, year, month, file)) as f:
                            items = f.read().split('\n')
                            for item in items:
                                if len(item) != 0:
                                    j = json.loads(item)
                                    headline = j['Headline']
                                    date = j['Date']
                                    content = j['News']
                                    if date:
                                        if content.startswith('Supported by'):
                                            content = content.replace('Supported by', '', 1)
                                        if content.startswith('AdvertisementSupported by'):
                                            content = content.replace('AdvertisementSupported by', '', 1)
                                        headlines.append(headline)
                                        dates.append(date)
                                        contents.append(content)

# Create Pandas Dataframe data.
NYT_df = pd.DataFrame({'headline': headlines,
     'date': dates,
     'content': contents,
    })
NYT_df
```

```python
# Combine two sources of news.
combined_df = pd.concat([Guardian_df, NYT_df])
combined_df
```

```python
# Combine news in 2019.
news_2019 = pd.DataFrame(columns=['headline','date','content'])
root = '../../Data/JsonNews/2019'
media_list = os.listdir(root)
# news_num = 0
for m in media_list:
    now_dir = os.path.join(root,m)
    if os.path.isdir(now_dir):
        for file in os.listdir(now_dir):
            if file.endswith('.pkl'):
                news = pd.read_pickle(os.path.join(now_dir,file))
#                 news_num += len(news)
                news_2019 = pd.concat([news_2019,news])
                
combined_df = pd.concat([combined_df,news_2019]) # combine news in 2019.
```

```python
# Preprocess headline and content.
import nltk
nltk.download('stopwords')
import string
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

def tokenize(text):
    if text:
        text = text.lower()
        tokens = RegexpTokenizer(r'\w+').tokenize(text)
        return tokens

def remove_stopwords(tokens):
    if tokens:
        removed = [i for i in tokens if not i in stop_words]
        return removed

def compute_length(tokens):
    if tokens:
        return len(tokens)

# combined_df['headline'] = combined_df['headline'].apply(tokenize_remove_stop)
combined_df['tokens'] = combined_df['content'].apply(tokenize)
combined_df['tokens_remove_stopwords'] = combined_df['tokens'].apply(remove_stopwords)
combined_df['content_length'] = combined_df['tokens_remove_stopwords'].apply(compute_length)
```

```python
# Plot distribution of length of news.
%matplotlib inline
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1, figsize=(20,10))

ax = plt.hist(combined_df['content_length'], bins=1000)
plt.xlabel('length of news')
plt.ylabel('count')
plt.title('Distribution of length of news')
plt.show()
```

```python
# Choose proper length of news for downstream work.
combined_df = combined_df[(combined_df['content_length'] > 20) & (combined_df['content_length'] < 1200)]
combined_df
```

```python
# Lemmatization.
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
lemmatizer = WordNetLemmatizer()

def lemantize(tokens):
    if tokens:
        return [lemmatizer.lemmatize(word) for word in tokens]

# combined_df['le_headline'] = combined_df['headline'].apply(lemantize)
combined_df['le_content'] = combined_df['tokens_remove_stopwords'].apply(lemantize)
```
