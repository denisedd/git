## Visualzie words embedding through t-SNE
1. read dataset of Nasdaq Index
2. dicover the stock price trend and choose one interval to visualzie the news healine during that interval <br>
** For instance, Nasdaq Index dropped significantly at the end of 2018. Therefore, I would like to know what happens during that period. **
3. Since there was a sharp decline on stock price during that period, I look for the negative news corresponding to that time interval.
4. First, I use wordcloud to see the most frquent words of negative news
5. Then, I get top 70 common words of negative news and fit them into t-SNE model with its reduce-dimensional method
6. Then, I can generate the plot to see how these most frequent words from negative news are related with each other

```python
import os
import json
import pandas as pd
import numpy as np
from pandas.io.json import json_normalize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.sentiment.util import *
import quandl

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# plot the Nasdaq Index value from 20080608 - 20190403
ndq = quandl.get("NASDAQOMX/COMP-NASDAQ",
                        start_date = '20080608', 
                        end_date = '20190403')
ndq_df = ndq.reset_index()
plt.plot(ndq_df['Trade Date'], ndq_df['Index Value'])
```

```python
# concat 2008 - 2019 dataset
# 2008-06-08
data = pd.read_csv('../../Data/stocknews/RedditNews.csv')
data['Date'] = pd.to_datetime(data['Date'])
df = data[data['Date'] < '20140101']
df['News'] = df['News'].str.strip("b'")
df = df.sort_values(by='Date')
df.rename(index=str, columns={"Date": "date", "News": "headline"}, inplace =True)

data_1 = pd.read_pickle('../../Data/ProcessedNews.pkl')
data_1['date'] = pd.to_datetime(data_1['date'])
df_2 = data_1.sort_values(by='date')
df_2 = df_2[['date', 'headline']]
df_2 = df_2.loc[df_2['date'] >= '20140101']
```

```python
# concat 2 dataframe from 2008-2019
df_3 = pd.concat([df, df_2], sort=False)
df_3.head()
```

```python
def concat_headline(df, col_name):
    if col_name == 'sentiment':
        result_1 = df[['date', 'sentiment']]
    elif col_name == 'proc_headline':
        result_1 = df[['date', 'proc_headline']]
    result_2 = result_1.groupby('date').cumcount() + 1
    result_3 = result_1.set_index(['date', result_2]).unstack().sort_index(1, level=1)
    result_3.columns = ['_'.join(map(str,i)) for i in result_3.columns]
    result_3 = result_3.reset_index()
    result_3['date'] = pd.to_datetime(result_3['date'])
    return result_3


df_3['headline'] = df_3['headline'].fillna(" ")
sid = SentimentIntensityAnalyzer()
df_3['sentiment'] = df_3['headline'].apply(lambda x:sid.polarity_scores(x)['compound'] if (len(x)>3) else float(0))
df_5 = concat_headline(df_3, 'sentiment')
df_5.head()
```

```python
def senti_mean(df_2):
    # calculate sentiment mean of headline from news every day
    ttl=[]
    for row in range(df_2.shape[0]):
        lst=[]
        for col in range(1, df_2.shape[1]):
            if (df_2.iloc[row, col] is not None) or (pd.notnull(df_2.iloc[row, col])):
                    lst.append(df_2.iloc[row, col].astype(float))
        cleanedList = [x for x in lst if str(x) != 'nan']
        arr_mean = np.mean(cleanedList, axis=0)
        ttl.append(arr_mean)
    return ttl

ttl = senti_mean(df_5)
df_5['average_senti'] = ttl
df_6 = df_5[['date', 'average_senti']]
df_6.head()
```

```python
# remove stopwords, tkenize, lemmatize
import nltk
import string
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def tokenize_remove_stop(text):
    if text:
        text = text.lower()
        tokens = RegexpTokenizer(r'\w+').tokenize(text)
        result = [i for i in tokens if not i in stop_words]
        return result
    
df_3['proc_headline'] = df_3['headline'].apply(tokenize_remove_stop)
```

```python
# see the negative news headline for t-SNE plot
df_7 = df_6.loc[(df_6['date'] >= '20080920') & (df_6['date'] <= '20081025') & (df_6['average_senti'] < 0)]
# df_8 = concat_headline(df_3, 'proc_headline')
df_8 = df_3[['date', 'proc_headline']]
aa = df_7.merge(df_8, how='left', on='date')
aa.head()
```

```python
headline = aa['proc_headline'].tolist()
# lst =[]
lst = [a for i in headline for a in i]
text = ' '.join(word for word in lst)
```

```python
from wordcloud import WordCloud
# Create and generate a word cloud image:
wordcloud = WordCloud(max_font_size=70, max_words=100, background_color="white").generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from collections import Counter

r = [x for x, _ in Counter(lst).most_common(70)]
text_lst = [word for word in r if len(word) >1]
```

```python
import gensim

# Load Google's pre-trained Word2Vec model.
model = gensim.models.KeyedVectors.load_word2vec_format('../../Data/GoogleNews-vectors-negative300.bin', binary=True)
```

```python
embedding_clusters = []
word_clusters = []
for word in lst:
    embeddings = []
    words = []
    try:
        for similar_word, _ in model.most_similar(word, topn=5):
            words.append(similar_word)
            embeddings.append(model[similar_word])
    except:
        continue
    embedding_clusters.append(embeddings)
    word_clusters.append(words)
    
```

```python
from sklearn.manifold import TSNE
import numpy as np

embedding_clusters = np.array(embedding_clusters)
n, m, k = embedding_clusters.shape
tsne_model_en_2d = TSNE(perplexity=15, n_components=2, init='pca', n_iter=2500, random_state=32)
embeddings_en_2d = np.array(tsne_model_en_2d.fit_transform(embedding_clusters.reshape(n * m, k))).reshape(n, m, 2)
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import warnings
warnings.filterwarnings('ignore')
% matplotlib inline
def tsne_plot(model, text_lst):
    "Creates and TSNE model and plots it"
    labels = []
    tokens = []

    for word in text_lst:
        tokens.append(model[word])
        labels.append(word)
    
    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
    new_values = tsne_model.fit_transform(tokens)

    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(7, 5)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')
    plt.show()
```

```python
# negative news from 20080920- 20081025
# only get top frequent 70 words
# and similar words are group together
tsne_plot(model, text_lst)
```
