# Baseline model for linear and binary prediction for nasdaq index
1. Simple Linear Regression model is defined to predict stock price next day.
2. Logistic Regression model is defined to predict binary movement next day.

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn import preprocessing
import quandl
```

```python
# Load stock price data from csv file.
def stocks(start_date, end_date):
# use quandl to acquire nasdaq composite
    ndq = quandl.get("NASDAQOMX/COMP-NASDAQ",
                        start_date = start_date, 
                        end_date = end_date)
    ndq_df = ndq.reset_index()
    return ndq_df

stockdf = stocks('2014-01-01', '2018-12-31')
# stockdf
# Preprocessing data.
stockdf['Trade Date'] = pd.to_datetime(stockdf['Trade Date'])
stockdf = stockdf[(stockdf['Trade Date'] >= pd.Timestamp(2016, 1, 1)) & (stockdf['Trade Date'] <= pd.Timestamp(2018, 12, 31))]
stockdf = stockdf.reset_index(drop=True)
# companydf['time_delta'] = (companydf['date'] - pd.Timestamp(2016, 1, 1)).map(lambda x: x.days)
# companydf['diff'] = companydf['5. adjusted close'].diff()
# companydf['diff'] = companydf['diff'].fillna(0.0)
# companydf['log_ret'] = np.log(companydf['5. adjusted close']) - np.log(companydf['5. adjusted close'].shift(1))
# companydf = companydf.iloc[1:].drop(['4. close', '7. dividend amount','8. split coefficient'], axis=1)
assert not stockdf.isnull().values.any()
stockdf
```

```python
# Normalizing data.
zero_one_scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
# neg_pos_scaler = preprocessing.MinMaxScaler(feature_range=(-1, 1))
# opn = zero_one_scaler.fit_transform(np.array(companydf['1. open']).reshape(-1,1))
# hih = zero_one_scaler.fit_transform(np.array(companydf['2. high']).reshape(-1,1))
# low = zero_one_scaler.fit_transform(np.array(companydf['3. low']).reshape(-1,1))
stockdf['normalized_index_value'] = zero_one_scaler.fit_transform(np.array(stockdf['Index Value']).reshape(-1,1))
# vol = zero_one_scaler.fit_transform(np.array(companydf['6. volume']).reshape(-1,1))
# log_ret = neg_pos_scaler.fit_transform(np.array(companydf['log_ret']).reshape(-1,1))
# date = np.array(companydf['date']).reshape(-1,1)
# normalized_data = np.concatenate((opn, hih, low, adj_cls, vol, log_ret), axis=1, out=None)

# normalized_df = pd.DataFrame(normalized_data, columns=['open', 'high', 'low', 'adjusted_close', 'volume', 'log_return'])
stockdf
```

```python
plt.figure(figsize=(20,10))
plt.plot(stockdf['Trade Date'], stockdf['Index Value'], color='blue', linewidth=4)
plt.xlabel('time')
plt.ylabel('price')
plt.title('Nasdaq Index')
```

```python
import sklearn
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

stockdf['target'] = stockdf['normalized_index_value'].shift(-1)
stockdf = stockdf.iloc[:-1]

train_data = stockdf.iloc[:int(len(stockdf)*0.8)]
test_data = stockdf.iloc[int(len(stockdf)*0.8):]

train_X = train_data[['normalized_index_value']].copy().values
train_Y = train_data[['target']].copy().values

test_X = test_data[['normalized_index_value']].copy().values
test_Y = test_data[['target']].copy().values

poly = sklearn.preprocessing.PolynomialFeatures(1)
train_X_t = poly.fit_transform(train_X)
test_X_t = poly.fit_transform(test_X)

linear_mod = linear_model.LinearRegression() # defining the linear regression model
linear_mod.fit(train_X_t, train_Y) # fitting the data points in the model

predict_train = linear_mod.predict(train_X_t)
predict_test = linear_mod.predict(test_X_t)

mse = mean_squared_error(test_Y,predict_test)
r2 = r2_score(test_Y,predict_test)
```

```python
stockdf['diff'] = stockdf['normalized_index_value'].diff()
stockdf['diff'] = stockdf['diff'].fillna(0.0)
stockdf['label'] = stockdf['diff'].map(lambda x : 1 if x >0 else 0)
stockdf['label'] = stockdf['label'].shift(-1)
stockdf = stockdf[:-1]
```

```python
stockdf
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegressionCV
from sklearn.linear_model import Ridge
from sklearn.model_selection import validation_curve

train_data = stockdf.iloc[:int(len(stockdf)*0.8)]
test_data = stockdf.iloc[int(len(stockdf)*0.8):]

train_X = train_data[['normalized_index_value']].copy().values
train_Y = train_data[['label']].copy().values.reshape(-1,)

test_X = test_data[['normalized_index_value']].copy().values
test_Y = test_data[['label']].copy().values.reshape(-1,)

model = LogisticRegressionCV(solver='lbfgs', cv = 10) 
model.fit(train_X, train_Y) 

predict_train = model.predict(train_X)
predict_test = model.predict(test_X)

acc = model.score(test_X, test_Y)
acc
```
