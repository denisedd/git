```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn import preprocessing
```

```python
# Load stock price data from csv file.
root = '../Data/Stock_price'
company = 'FB'
companydf = pd.read_csv(os.path.join(root, company + '.csv'))

# Preprocessing data.
companydf['date'] = pd.to_datetime(companydf['date'])
companydf = companydf[(companydf['date'] >= pd.Timestamp(2016, 1, 1)) & (companydf['date'] <= pd.Timestamp(2018, 12, 31))]
companydf = companydf.reset_index(drop=True)
# companydf['time_delta'] = (companydf['date'] - pd.Timestamp(2016, 1, 1)).map(lambda x: x.days)
# companydf['diff'] = companydf['5. adjusted close'].diff()
# companydf['diff'] = companydf['diff'].fillna(0.0)
companydf['log_ret'] = np.log(companydf['5. adjusted close']) - np.log(companydf['5. adjusted close'].shift(1))
companydf = companydf.iloc[1:].drop(['4. close', '7. dividend amount','8. split coefficient'], axis=1)
assert not companydf.isnull().values.any()
companydf
```

```python
# Normalizing data.
zero_one_scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
neg_pos_scaler = preprocessing.MinMaxScaler(feature_range=(-1, 1))
opn = zero_one_scaler.fit_transform(np.array(companydf['1. open']).reshape(-1,1))
hih = zero_one_scaler.fit_transform(np.array(companydf['2. high']).reshape(-1,1))
low = zero_one_scaler.fit_transform(np.array(companydf['3. low']).reshape(-1,1))
adj_cls = zero_one_scaler.fit_transform(np.array(companydf['5. adjusted close']).reshape(-1,1))
vol = zero_one_scaler.fit_transform(np.array(companydf['6. volume']).reshape(-1,1))
log_ret = neg_pos_scaler.fit_transform(np.array(companydf['log_ret']).reshape(-1,1))
# date = np.array(companydf['date']).reshape(-1,1)
normalized_data = np.concatenate((opn, hih, low, adj_cls, vol, log_ret), axis=1, out=None)

normalized_df = pd.DataFrame(normalized_data, columns=['open', 'high', 'low', 'adjusted_close', 'volume', 'log_return'])
normalized_df
```

```python
plt.figure(figsize=(20,10))
plt.plot(normalized_df.index, normalized_df['adjusted_close'], color='blue', linewidth=4)
plt.xlabel('time')
plt.ylabel('price')
plt.title(company)
```

```python
import sklearn
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

normalized_df['target'] = normalized_df['log_return'].shift(-1)
normalized_df = normalized_df.iloc[:-1]

train_data = normalized_df.iloc[:int(len(normalized_df)*0.8)]
test_data = normalized_df.iloc[int(len(normalized_df)*0.8):]

train_X = train_data.drop(['target'], axis=1).values
train_Y = train_data[['target']].copy().values

test_X = test_data.drop(['target'], axis=1).values
test_Y = test_data[['target']].copy().values

poly = sklearn.preprocessing.PolynomialFeatures(1)
train_X_t = poly.fit_transform(train_X)
test_X_t = poly.fit_transform(test_X)

linear_mod = linear_model.LinearRegression() # defining the linear regression model
linear_mod.fit(train_X_t, train_Y) # fitting the data points in the model

predict_train = linear_mod.predict(train_X_t)
predict_test = linear_mod.predict(test_X_t)

rmse = np.sqrt(mean_squared_error(test_Y,predict_test))
r2 = r2_score(test_Y,predict_test)
```

```python
from pandas.tseries.offsets import DateOffset
fig = plt.figure(figsize=(20,10))
ax = plt.subplot(111)

ax.plot(normalized_df.index, normalized_df['adjusted_close'], color='blue', linewidth=1.5, alpha=1, label='Target')
ax.plot(train_data.index[1:], predict_train[:-1], color='orange', linewidth=1.5, alpha=1, label='Train')
ax.plot(test_data.index[1:], predict_test[:-1], color='green', linewidth=1.5, alpha=1, label='Test')

plt.xlabel('date')
plt.ylabel('price')
plt.title(company, fontsize=20)
ax.legend(prop={'size': 20})

print('rmse: ', rmse)
print('r2: ', r2)
```
